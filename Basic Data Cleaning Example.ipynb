### Empty and missing values
The most striking aspect of this dataset are the missing values for the "Company" feature

```python
Company_Null = hotelbooking_data['company'].isnull().sum()
Company_NotNull = hotelbooking_data['company'].notnull().sum()
Total_Company = np.product(hotelbooking_data['company'].shape)
```

```python
Company_NotNull
```

```python
Total_Company
```

Here we can see that a vast amount of the data in the feature "company" is missing meaning it will not be of much use for machine learning models which lead me to the decision of dropping it from the dataset.

```python
Company_Null/Total_Company *100
```

### Removing unnecessary features. 

The following features basically contain the same information is duplicated
#### "arrival_date_year,	arrival_date_month,	arrival_date_day_of_month" and "reservation_status_date" 
 

Using pandas to_datetime which allows use to transform the data time from string to a Date time object which would be more effective when it comes to the analysis

```python
hotelbooking_data.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'],axis =1,inplace =True)
```

```python
hotelbooking_data.head()
```

```python
hotelbooking_data['reservation_status_date'] = pd.to_datetime(hotelbooking_data['reservation_status_date'])
```
